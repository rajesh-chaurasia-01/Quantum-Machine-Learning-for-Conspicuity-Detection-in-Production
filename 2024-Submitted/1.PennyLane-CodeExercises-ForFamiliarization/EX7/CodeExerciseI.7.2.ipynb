```python
import numpy as np 
import random
import pennylane as qml
import matplotlib.pyplot as plt

dev = qml.device("default.qubit", wires=1)

@qml.qnode(dev)
def convert_to_rz_rx():
    ##################
    # IMPLEMENT THE CIRCUIT IN THE PICTURE USING ONLY RZ AND RX
    #NB: STdagger = RZ( pi/2 + pi/4 ) =RZ( 3pi/4 )
    # H = RZ(pi/2)RX(pi)RZ(0)
    # Y = RZ(pi)RX(pi)
    pi = np.pi
    qml.RZ(pi/2,wires=0)
    qml.RX(pi,wires=0) 
    #qml.RZ(0,wires=0)
    qml.RZ(pi/4,wires=0) 
    qml.RZ(pi,wires=0)   
    qml.RX(pi,wires=0) 
    ##################  

    return qml.state()
        
circuit = qml.QNode(convert_to_rz_rx, dev)
qml.drawer.use_style("pennylane")
result = qml.draw_mpl(circuit)()
plt.show()
```
